```python
med1 = pd.read_csv("query_outputs/medical/medicine/Animalia/specimen/specimen_medicine_animalia_trades_per_year_by_app.csv")
# would have been easier to have queried with a group by for purposes of visualization (seaborn like long data)
# but wanted to show that I could use case statements
# so now have to transform wide to long data using pd melt
```

```python
fig, axes1 = plt.subplots(1,3, figsize=(18,5))

fig.suptitle("Medicine Animalia Number of Species Trades vs Year by Appendix", size=20)
fig.subplots_adjust(top=.85)

appendices = ["I","II","III"]
colors = sns.color_palette("Set2",3)

for (num,app) in enumerate(appendices):
    cur_ax = axes1[num]
    sns.lineplot(data=med1, x="year", y=app, ax=cur_ax, color=colors[num])
    cur_ax.set_title(f"Appendix {app}", size=15)
    cur_ax.set_ylabel("")
    plt.ticklabel_format(style='plain', axis='y')

axes1[0].set_ylabel("Number of Specimens Traded", size=15)

fig.savefig("query_outputs/medical/medicine/Animalia/specimen")
```

```python
def app_graphs(title, dataframe):
    """Creates a set of trade graphs separated by appendix with data from a given dataframe
    
    Args:
        title (str): title of the graph
        dataframe: (pandas.DataFrame): dataframe with year and columns for each appendix
    """
    
    fig, axs = plt.subplots(1,3)
    
    fig.suptitle(title, size=20)
    fig.subplots_adjust(top=.85)
    
    appendices = ["I","II","III"]
    colors = sns.color_palette("Set2",3)

    for (num,app) in enumerate(appendices):
        cur_ax = axs[num]
        sns.lineplot(data=dataframe, x="year", y=app, ax=cur_ax, color=colors[num])
        cur_ax.set_title(f"Appendix {app}", size=15)
        cur_ax.set_ylabel("")
        plt.ticklabel_format(style='plain', axis='y')
    
    axs[0].set_ylabel("Number of Specimens Traded", size=15)
```
